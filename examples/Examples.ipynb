```python
%%cython
cimport cyrk4._cyrk4 as c4
cimport numpy as np
import numpy as np

def get_gaussians(N, sigma):
    cdef np.ndarray ret = np.zeros(N)
    c4.sample_gaussian(<double*> np.PyArray_DATA(ret), N, sigma)
    return ret
```

## Test ODE and SDE integrators

```python
%%cython -lm 

cimport cyrk4._cyrk4 as c4
cimport numpy as np
import numpy as np
from libc.math cimport sqrt

cdef struct my_params:
    double omega
    double kappa
    double epsilon
    

cdef void my_ode(double t, double * y, double * ydot, void * params):
    cdef my_params * p = <my_params*> params
    
    ydot[0] = -p[0].kappa/2 * y[0] + p[0].omega * y[1]
    ydot[1] = -p[0].omega * y[0] - p[0].kappa/2 * y[1] - sqrt(p[0].kappa) * p[0].epsilon
    
    
cdef void my_sde(double t, double * y, double * w, double * ydot, void * params):
    cdef my_params * p = <my_params*> params
    ydot[0] = -p[0].kappa/2 * y[0] + p[0].omega * y[1] - sqrt(p[0].kappa) * w[0]/2
    ydot[1] = -p[0].omega * y[0] - p[0].kappa/2 * y[1] - sqrt(p[0].kappa) * (w[1]/2 + p[0].epsilon)

def solve_my_ode(y0, tlist, hmax=.005, omega=1., kappa=1., epsilon=10.):
    cdef my_params p
    p.omega = omega
    p.kappa = kappa
    p.epsilon = epsilon
    return c4.integrate_ode_c_simple(my_ode, np.ascontiguousarray(tlist, np.float64), np.ascontiguousarray(y0, np.float64), hmax, &p)

def solve_my_sde(y0, tlist, hmax=.005, omega=1., kappa=1., epsilon=10.):
    cdef my_params p
    p.omega = omega
    p.kappa = kappa
    p.epsilon = epsilon
    return c4.integrate_sde_c_simple(my_sde, np.ascontiguousarray(tlist, np.float64), np.ascontiguousarray(y0, np.float64), hmax, 2, &p)

```

```python
times = linspace(0, 10, 2001)
y0 = zeros(2)
data1 = solve_my_ode(y0, times, hmax = .0001)
data2, noise = solve_my_sde(y0, times, hmax = 0.0001)
```

## Trajectories
